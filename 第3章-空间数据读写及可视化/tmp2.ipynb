```python
import rasterio as rio 
import matplotlib.pyplot as plt
```

```python
path_data = 'data/section-3/s2_20200411_6bands_chenggong.tif'
```

```python
rsimg = rio.open(path_data)
rsimg_arr = rsimg.read()
rsimg_arr.shape 
```

```python
fig, ax = plt.subplots(1,1,figsize=(5,5))
band1 = rsimg_arr[0,:,:]
band1.shape
ax.imshow(band1) 
```

```python
fig, ax = plt.subplots(1,1,figsize=(5,5))
ax.hist(band1.flatten(), bins=100, 
        range=[0,2000]
        );
```

```python
fig, ax = plt.subplots(1,1,figsize=(5,5))
ax.imshow(band1_) 
```

```python
band_cor = rsimg_arr[0:3,:,:].transpose(1,2,0)
band_cor.shape
```
