```python
import matplotlib.pyplot as plt
import numpy as np
```

线性图

```python
x = np.linspace(1, 10, 10)
# y = np.random.randn(10)
# y = 1*x + 2
y = x**2 + 2
fig, ax = plt.subplots(1,1)
# ax.plot(x, y)
# ax.plot(x, y, '.')
# ax.plot(x, y, '.', color='r', markersize=12)  ## 设置颜色，点大小
ax.plot(x, y, color='r', linewidth=2)   ## 设置线宽
```

设置坐标轴

```python
subjects = ['math', 'physics', 'english', 'geograpy']
scores = [98,67,86,79]
fig, ax = plt.subplots(1,1)
ax.plot(subjects, scores, color='r', linewidth=2)   ## 设置线宽度。
ax.set_xlabel('Subjects', fontsize=20)
ax.set_ylabel('Scores', fontsize=20)
ax.set_xlim(-0.4,3.4)
ax.set_ylim(60, 105)
ax.set_yticks([60, 80, 100])
ax.set_yticks([60, 80, 100], ['bad', 'good', 'perfect'])
ax.grid(True)
ax.tick_params(axis='x', labelsize=16, labelcolor='blue')
ax.tick_params(axis='y', direction='in')

```

```python
subjects = ['math', 'physics', 'english', 'geograpy']
zhang3 = [98,67,86,79]
li4 = [82,77,69,92]
fig, ax = plt.subplots(1,1)
ax.plot(subjects, zhang3, color='blue', linewidth=2, label='zhang3')   ## 设置线宽度。
ax.plot(subjects, li4, color='r', linewidth=2, label='li4')   ## 设置线宽度。
# ax.legend()
ax.legend(loc='lower left', title="student")

```
