#### 展示选课同学地理信息
1. 显示全国矢量地图
2. 显示同学家乡地理位置；
3. 在同学家乡位置上显示同学名字

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['font.sans-serif'] = ['SimHei']
```

```python
path_china = 'data/data-base/ChinaAdminDivisonSHP/2. Province/province.shp'
path_stud = 'students/students_hometown.csv'
china_gpd = gpd.read_file(path_china)
stud_info = pd.read_csv(path_stud)
```

```python
### 1. 显示地图
fig, ax = plt.subplots(figsize=(12, 8))
ax = china_gpd.plot(ax=ax, facecolor='None', edgecolor='black', linewidth=0.5)
### 2. 显示同学家乡位置
ax.scatter(stud_info['家乡经度'], stud_info['家乡纬度'], color='red', s=20)
### 3. 显示同学名字
for x, y, label in zip(stud_info['家乡经度'], stud_info['家乡纬度'], stud_info['姓名']):
    ax.text(x, y, label, fontsize=9, color='green')
```
